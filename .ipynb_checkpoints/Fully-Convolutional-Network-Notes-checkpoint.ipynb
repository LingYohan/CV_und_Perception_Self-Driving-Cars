## VGGnet architecture for image classification

#### We are going to modify VGGnet to better suite our needs, which in this case is segmentation. Here the input and the output share the same dimensions. VGGnet is composed of convolutional blocks placed sequentially to each other, each block consists of 2 convolutional layer followed by a max pooling layer.

#### We care about extracting the output of the last 3 blocks, i.e, pool3, pool4 and pool5. We upsample pool5 (x2) giving us a bigger image. We add this image to the output of pool4. Then, we upsample this summed image 2x, and then we add this image to the output of pool3. Finally, this summed image is upsampled 8x. Tihs final output is our segmentation mask. This is the FCN-8 architecture. (there are FCN-16, FCN-32 etc). 
#### FCN-32 identifies more coarse structures in images, whereas FCN-16 identifies more finer structures. FCN-8 identifies even more finer structures in the image.
